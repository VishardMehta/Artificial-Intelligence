```python
import copy

class Puzzle:
    def __init__(self, start, goal):
        self.curr = start
        self.goal = goal
        self.zero = start.index(0)  # Find empty tile (0)

    def print_state(self):
        for i in range(3):
            print(self.curr[i * 3:(i + 1) * 3])
        print()

    def move(self, direction):
        new_state = self.curr[:]
        if direction == "up" and self.zero > 2:
            new_state[self.zero], new_state[self.zero - 3] = new_state[self.zero - 3], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        if direction == "down" and self.zero < 6:
            new_state[self.zero], new_state[self.zero + 3] = new_state[self.zero + 3], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        if direction == "left" and self.zero % 3 > 0:
            new_state[self.zero], new_state[self.zero - 1] = new_state[self.zero - 1], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        if direction == "right" and self.zero % 3 < 2:
            new_state[self.zero], new_state[self.zero + 1] = new_state[self.zero + 1], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        return None

    def heuristic(self):
        return sum(1 for i in range(9) if self.curr[i] != self.goal[i])  # Count misplaced tiles

def hill_climb(p):
    closed = []
    while True:
        p.print_state()
        if p.curr == p.goal:
            print("Goal Reached!")
            return
        closed.append(p.curr)

        moves = [p.move(d) for d in ["up", "down", "left", "right"]]
        moves = [m for m in moves if m and m.curr not in closed]

        if not moves:
            print("Local Maxima! No better move found.")
            return

        p = min(moves, key=lambda x: x.heuristic())  # Pick best move

# Initial and Goal States
start_state = [1, 2, 3, 0, 4, 5, 6, 8, 7]  # Given start
goal_state = [2, 8, 1, 0, 4, 3, 7, 6, 5]   # Given goal

puzzle = Puzzle(start_state, goal_state)
hill_climb(puzzle)
```

```python
import copy

class Puzzle:
    def __init__(self, start, goal):
        self.curr = start
        self.goal = goal
        self.zero = start.index(0)

    def print_state(self):
        for i in range(3):
            print(self.curr[i * 3:(i + 1) * 3])
        print()

    def move(self, direction):
        new_state = self.curr[:]
        if direction == "up" and self.zero > 2:
            new_state[self.zero], new_state[self.zero - 3] = new_state[self.zero - 3], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        if direction == "down" and self.zero < 6:
            new_state[self.zero], new_state[self.zero + 3] = new_state[self.zero + 3], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        if direction == "left" and self.zero % 3 > 0:
            new_state[self.zero], new_state[self.zero - 1] = new_state[self.zero - 1], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        if direction == "right" and self.zero % 3 < 2:
            new_state[self.zero], new_state[self.zero + 1] = new_state[self.zero + 1], new_state[self.zero]
            return Puzzle(new_state, self.goal)
        return None

    def heuristic(self):
        return sum(1 for i in range(9) if self.curr[i] != self.goal[i])

def steepest_hill_climb(p):
    closed = []
    while True:
        p.print_state()
        if p.curr == p.goal:
            print("Goal Reached!")
            return
        closed.append(p.curr)

        moves = [p.move(d) for d in ["up", "down", "left", "right"]]
        moves = [m for m in moves if m and m.curr not in closed]

        if not moves:
            print("Local Maxima! No better move found.")
            return

        best_moves = sorted(moves, key=lambda x: x.heuristic())
        best_move = best_moves[0]

        if best_move.heuristic() >= p.heuristic():
            print("No better move available. Stuck at local maxima!")
            return

        p = best_move


start_state = [1, 2, 3, 0, 4, 5, 6, 8, 7]
goal_state = [2, 8, 1, 0, 4, 3, 7, 6, 5]

puzzle = Puzzle(start_state, goal_state)
steepest_hill_climb(puzzle)
```

```python
class WaterJug:
    def __init__(self):
        self.start = (0, 0)
        self.goal = (2, _)

    def moves(self, state):
        x, y = state
        possible_moves = [
            (4, y),
            (x, 3),
            (0, y),
            (x, 0),
            (max(0, x - (3 - y)), min(3, y + x)),
            (min(4, x + y), max(0, y - (4 - x)))
        ]
        return possible_moves

    def heuristic(self, state):
        return abs(state[0] - 2)

    def simple_hill_climb(self):
        print("\n--- Simple Hill Climb ---")
        state = self.start
        visited = []

        while True:
            print(state)
            if state[0] == 2:
                print("Goal reached!")
                return
            visited.append(state)

            for move in self.moves(state):
                if move not in visited and self.heuristic(move) < self.heuristic(state):
                    state = move
                    break
            else:
                print("Stuck at local maxima!")
                return

    def steepest_hill_climb(self):
        print("\n--- Steepest Hill Climb ---")
        state = self.start
        visited = []

        while True:
            print(state)
            if state[0] == 2:
                print("Goal reached!")
                return
            visited.append(state)

            next_moves = [m for m in self.moves(state) if m not in visited]
            if not next_moves:
                print("Stuck at local maxima!")
                return

            state = min(next_moves, key=self.heuristic)

jug_problem = WaterJug()
jug_problem.simple_hill_climb()
jug_problem.steepest_hill_climb()
```
