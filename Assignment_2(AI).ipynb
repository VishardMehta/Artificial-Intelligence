```python
from collections import deque

class jugProblem:
    def __init__(self, jug1, jug2, jug1final, jug2final):
        self.jug1_capacity = jug1
        self.jug2_capacity = jug2
        self.jug1_final = jug1final
        self.jug2_final = jug2final
        self.visited = set()  # no revisiting

    def is_goal(self, state):
        return state == (self.jug1_final, self.jug2_final)

    def get_next_states(self, state):
        jug1, jug2 = state
        next_states = []

        # Fill 1
        next_states.append((self.jug1_capacity, jug2))
        # Fill 2
        next_states.append((jug1, self.jug2_capacity))
        # Empty 1
        next_states.append((0, jug2))
        # Empty 2
        next_states.append((jug1, 0))
        # 1 -> 2
        pour = min(jug1, self.jug2_capacity - jug2)
        next_states.append((jug1 - pour, jug2 + pour))
        # 2 -> 1
        pour = min(jug2, self.jug1_capacity - jug1)
        next_states.append((jug1 + pour, jug2 - pour))

        return [s for s in next_states if s not in self.visited]

    def bfs(self):
        queue = deque([((0, 0), [])])
        self.visited.clear()

        while queue:
            (jug1, jug2), path = queue.popleft()
            if self.is_goal((jug1, jug2)):
                return path + [(jug1, jug2)]

            self.visited.add((jug1, jug2))
            for next_state in self.get_next_states((jug1, jug2)):
                queue.append((next_state, path + [(jug1, jug2)]))

        return "No solution found"

    def dfs(self, state=(0, 0), path=[]):
        if self.is_goal(state):
            return path + [state]

        self.visited.add(state)
        for next_state in self.get_next_states(state):
            if next_state not in self.visited:
                result = self.dfs(next_state, path + [state])
                if result:
                    return result

        return None

j = jugProblem(4, 3, 2, 0)

print("Jug 1 is 4l , Jug 2 is 3l ")
print("\n")
# Solve using BFS
print("BFS Solution:")
bfs_solution = j.bfs()
if bfs_solution:
    for step in bfs_solution:
        print(f"Jug 1: {step[0]}L, Jug 2: {step[1]}L")
else:
    print("No solution found.")

# Solve using DFS
print("\nDFS Solution:")
j.visited.clear()
dfs_solution = j.dfs()
if dfs_solution:
    for step in dfs_solution:
        print(f"Jug 1: {step[0]}L, Jug 2: {step[1]}L")
else:
    print("No solution found.")
```

```python
import numpy as np
from itertools import permutations

graph = [
    [0, 10, 15, 20],  #  1
    [10, 0, 35, 25],  #  2
    [15, 35, 0, 30],  #  3
    [20, 25, 30, 0]   #  4
]

num_nodes = len(graph)

start = int(input("Enter starting node (1-4): ")) - 1

def tsp_bruteforce(graph, start):
    nodes = list(range(num_nodes))
    nodes.remove(start)

    min_cost = float('inf')
    best_path = None

    for perm in permutations(nodes):
        current_cost = 0
        current_path = [start] + list(perm) + [start]

        for i in range(num_nodes):
            current_cost += graph[current_path[i]][current_path[i+1]]

        if current_cost < min_cost:
            min_cost = current_cost
            best_path = current_path

    return best_path, min_cost

path, cost = tsp_bruteforce(graph, start)

print(f"Optimal Path: {' -> '.join(str(p+1) for p in path)}")
print(f"Minimum Cost: {cost}")
```
